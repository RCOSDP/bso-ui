```python
import requests
```

```python
url = "http://localhost:9200/bsso-publications/_search"
```

```python
year = "2021Q1" # et aussi tous les "mill√©simes" 2018, 2019, 2020
params = {
  "aggs": {
    "by_publication_year": {
      "terms": {
        "field": "publication_year"
      },
      "aggs": {
        "by_is_oa": {
          "terms": {
            "field": "oa_details." + year + ".is_oa"
          }
        }
      }
    }
  }
}

requests.get(url, json=params).json()['aggregations']
```
